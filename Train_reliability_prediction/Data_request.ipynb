## Importing libraries

```python
from bs4 import BeautifulSoup
import requests
import cloudscraper
import pandas as pd
```

## Pull requests
Avoid cloudfare protection against bots

```python
url = 'https://www.cleartrip.com/trains/stations/SC/'

scraper = cloudscraper.create_scraper() # for scraping from cloudfare protected sites
page = scraper.get(url)
# print(page)

soup = BeautifulSoup(page.text, "html")
```

Pulls out html text

```python
print(soup)
```

```python
table = soup.find_all('table', class_ = 'results')[0]
```

```python
print(table)
```

```python
titles = table.find_all('th')
```

```python
print(titles)
```

```python
rail_titles = [title.text for title in titles]
print(rail_titles)
```

### Convert into dataframes

```python
df = pd.DataFrame(columns = rail_titles)

df
```

```python
columns_data = table.find_all('tr')
columns_data
```

Adding subsequent rows

```python
for row in columns_data[1:]:
  row_data = row.find_all('td')
  entries = [entry.text for entry in row_data]
  length = len(df)
  df.loc[length] = entries

df.head()
```

```python
print(df.columns)
```

```python
train_no = [train.split("(") for train in df["Train name (no.) passing via  Secunderabad Junction"]]
```

```python
# train_no
numbers = [number[1].split(')') for number in train_no]
numbers = [number[0] for number in numbers]

train_num = pd.DataFrame(numbers, columns = ["Train number"])

final_df = pd.concat([train_num, df], axis = 1)
final_df.head()
```

### Save the file in csv format !!

```python
final_df.to_csv(r"sample_data/train_schedule_SC.csv")
```

## Scraping average delay in arrival times for certain trains

_..... Starting of a repetitive block of code to accumulate all data from a sequence of webpages_

```python
url2 = "https://www.railyatri.in//insights/average-train-delay-at-station/sc"

page2 = requests.get(url2)

soup2 = BeautifulSoup(page2.text, "html")

print(soup2)
```

- **This snippet ensures that the we can navigate to the next webpage**

```python
next = soup2.find_all('div', class_ = 'pages')[0]

to = next.find_all('a')[1]

to
```

```python
delay_board = soup2.find_all('script', type = 'text/javascript')[2]

print(delay_board)
```

```python
stage = delay_board.text

print(stage)
```

```python
pecker = stage.split('\n')

train_list = pecker[2]
```

```python
list_s_idx = train_list.find('[')

list_e_idx = train_list.find(']')

train_list = train_list[list_s_idx+1:list_e_idx]

print(train_list)
```

```python
delay_list = train_list.split('},{')

delay_list[0] = delay_list[0][1:]
delay_list[-1] = delay_list[-1][:-1]
```

```python
delay_list = ["{" + train_data + "}" for train_data in delay_list]

delay_list
```

- **This snippet is to be executed only for the first execution of the repetitive block**

```python
# delay_data = [eval(train_data) for train_data in delay_list]

# len(delay_data)
```

- **This snippet is to be executed for the further executions of the repetitive block**

```python
temp = [eval(train_data) for train_data in delay_list]

delay_data = delay_data + temp

len(delay_data)
```

_... The repetitive block ends here_

........

Now we will retrieve everything into DataFrame

```python
avg_delay = []

for dict in delay_data:
  if dict["number"] not in avg_delay:
    avg_delay.append(dict["number"])

tr_no = [number.split(' (')[0] for number in avg_delay]
delay_mins = [int(number.split(' (')[1].strip()[:-6]) for number in avg_delay]

print(delay_mins) #remember this is an int
```

```python
print(len(tr_no), len(delay_mins))
```

```python
# fetch delay for each of the 49 pointers
pointers = [];
temp = [];
last = delay_data[0]["number"]

for i in range(len(delay_data)):
  temp.append(delay_data[i]["value"])

  if delay_data[i]["number"] != last:
    pointers.append(temp)
    temp = []
    last = delay_data[i]["number"]

if len(temp) > 0:
  pointers.append(temp)

pointers = [tuple(ele) for ele in pointers]
```

```python
len(pointers)
```

```python
columns = ["ptr "+ str(i) for i in range(75)]

delay_frame = pd.DataFrame(pointers, columns = columns)
```

```python
delay_frame_2 = pd.DataFrame(zip(tr_no, delay_mins), columns = ["Train number", "Delay in mins"])
```

```python
delay_frame = pd.concat([delay_frame_2, delay_frame], axis = 1)
```
